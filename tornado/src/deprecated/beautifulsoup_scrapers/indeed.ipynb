```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json
```

```python
base_url = 'https://www.indeed.com/'
indeed_url = f'{base_url}jobs?q=software%20engineer&l=New%20York%2C%20NY&vjk=fcc61cb83ef048a8&advn=8589170693800026'

res = requests.get(indeed_url)
dom = BeautifulSoup(res.text,"html.parser")
```

```python
selects = dom.select("li")

prefer = "profi"

for select in selects:
    if prefer in select.get_text():
        print(select.get_text())
```
